```python
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
from cartopy.io.shapereader import Reader
from cartopy.feature import ShapelyFeature
import cartopy.feature as cfeature
import geopandas as gpd
```

```python
gdf = gpd.read_file('./muni_2018gw/muni_2018gw.shp')
gdf[gdf.NOM_ENT == "Campeche"].to_file("campeche.shp")
```

```python
campeche = 'campeche.shp'
estados = "./destdv250k_2gw/destdv250k_2gw.shp"

plt.figure(figsize=(10,10))
ax = plt.axes(projection=ccrs.PlateCarree())
ax.set_extent([-93,-88,16,23])
campeche = ShapelyFeature(Reader(campeche).geometries(), ccrs.PlateCarree(), facecolor='none')
estados = ShapelyFeature(Reader(estados).geometries(), ccrs.PlateCarree(), facecolor='none')

ax.add_feature(campeche)
ax.add_feature(estados)
ax.add_feature(cfeature.LAND)
ax.add_feature(cfeature.OCEAN)
#ax.add_feature(cfeature.BORDERS)


cpe_lon, cpe_lat = -90.52, 19.84

ax.annotate('Campeche\nLat:-90.52\nLon=19.84', xy=(cpe_lon, cpe_lat),
           xytext=(cpe_lon-1, cpe_lat+1), 
            arrowprops=dict(facecolor='black', arrowstyle='->'),
           bbox=dict(boxstyle="square", fc="white", ec="black", lw=2)
           )
ax.plot(cpe_lon, cpe_lat, 'ro', markersize=3, transform=ccrs.PlateCarree())
plt.show()
```
